```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.jaquet-droz.com/en/watches'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')

links = []
for i in range(0, 8):
    span = soup.find_all('span', class_='field-content')
    a = span[i].find('a')
    href = a['href']
    links.append(href)

data = {}

for link in links:
    time_url = link
    time_page = requests.get(time_url)
    time_soup = BeautifulSoup(time_page.text, 'html.parser')
    li = time_soup.find_all('li', class_='views-row')

    href_list = []
    src_list = []
    field_title_list = []

    for item in li:
        a_tag = item.find('a')
        href = a_tag['href']
        href_list.append(href)

        img_tag = item.find('img')
        src = img_tag['src']
        src_list.append(src)

        field_title = item.find('div', class_='field_title').text.strip()
        field_title_list.append(field_title)

    data[link] = {
        'href_list': href_list,
        'src_list': src_list,
        'field_title_list': field_title_list
    }

# Accessing the data for a specific link
example_link = links[0]
example_data = data[example_link]
example_href_list = example_data['href_list']
example_src_list = example_data['src_list']
example_field_title_list = example_data['field_title_list']

print("Example Href List:", example_href_list)
print("Example Src List:", example_src_list)
print("Example Field Title List:", example_field_title_list)
```
