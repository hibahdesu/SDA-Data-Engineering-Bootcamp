<h1><center>File Reading (Practical Example)</center></h1>



```python
'''
Name, Age, Grade
Alice, 25, A
Bob, 30, B
Charlie, 22, C
David, 28, A
Eva, 35, B
'''
```

### Reading Data from the File:

```python
# Read the contents of the file
with open('data.txt', 'r') as file:
    data = file.read()

print(data)
```

### Parsing Data into Lists:

```python
# Split the lines into a list
lines = data.split('\n')

# Remove the header and empty last line
header = lines[0].split(', ')
lines = [line.split(', ') for line in lines[1:-1]]

print("Header:", header)
print("Data:", lines)
```

### Filtering Data Using if/else:

```python
# Filter people with age greater than 25
filtered_data = [line for line in lines if int(line[1]) > 25]

print("People with Age > 25:", filtered_data)
```

### Updating Data:

```python
# Update Charlie's age to 23
for line in lines:
    if line[0] == 'Charlie':
        line[1] = '23'

print("Updated Data:", lines)
```

### Creating New Data:

```python
# Add a new person to the data
new_person = ['Frank', '27', 'B']
lines.append(new_person)

print("Updated Data with New Person:", lines)
```

```python
def is_prime(num):
    '''
    Naive method of checking for primes.
    '''
    for n in range(2,num):
        if num % n == 0:
            print('not prime')
            break
    else: # If never mod zero, then prime
        print('prime')
```

```python
is_prime(6)
```

```python
PI = 3.14159265359
def circumference_of_circle(radius):
    '''
    Return the circumference of the argument
    argument: radius
    '''
    circumference = 2 * PI * radius
    return circumference

circumference_of_circle(5)
```

```python
# function definition

def calc_standard_deviation(num_list):
    '''
    Return the standard deviation of the argument
    argument: list of number
    '''
    n = len(num_list)
    mean = sum(num_list) / n
    squared_distance = [((num - mean)**2) for num in num_list]
    return (sum(squared_distance) / n) ** 0.5

# Mean and STD of spring cohort students

spring_cohort = [65, 87, 83, 53, 72, 78, 72, 78, 82, 73, 76, 74, 81]

calc_standard_deviation(spring_cohort)
```

```python
def predict_fraud(amount_usd, card_country, countries_used):
    '''
    Return the fraud score
    argument: amount usd, card country, countries used
    '''
    if amount_usd > 20.00:
        if card_country == 'CA':
            fraud_score = 1.00
        else:
            fraud_score = 0.50
    else:
        if countries_used > 2:
            fraud_score = 0.25
        else:
            fraud_score = 1.00
    return fraud_score

fraud_score = predict_fraud(1.50, 'CA', 3)

print(fraud_score)
```

```python
def predict_fraud(amount_usd, countries_used, card_country='CA'):
    if amount_usd > 20.00:
      if card_country == 'CA':
          fraud_score = 1.00
      else:
         fraud_score = 0.50
    else:
      if countries_used > 2:
         fraud_score = 0.25
      else:
         fraud_score = 1.00
    return fraud_score

fraud_score = predict_fraud(amount_usd=25, countries_used=3)
print(fraud_score)
```

```python
# function definition
def calc_standard_deviation(*nums):
   n = len(nums)
   mean = sum(nums) / n
   squared_distance = [((num - mean)**2) for num in nums]
   return (sum(squared_distance) / (n - 1)) ** 0.5

# Mean and STD of spring cohort students
print(calc_standard_deviation(65, 87, 83))
print(calc_standard_deviation(65, 87, 83, 53, 72, 78))

spring_cohort = [65, 87, 83, 53, 72, 78, 72, 78, 82, 73, 76, 74, 81]
print(calc_standard_deviation(*spring_cohort))
```
