```python
2<10 and 2!=10
```

```python
(2<10) & (2!=10)
```

```python
'''
First, it evaluates 10 & 2, which is a bitwise AND operation between the
binary representations of 10 and 2. In binary, 10 is 1010 and 2 is 0010.
Performing a bitwise AND operation results in 0000, which is 0 in decimal.
Then, it evaluates 2 < 0, which is false.
'''
2<10 & 2!=10
```

<h1><center>File Reading (Practical Example)</center></h1>



```python
'''
Name, Age, Grade
Alice, 25, A
Bob, 30, B
Charlie, 22, C
David, 28, A
Eva, 35, B
'''
```

### Reading Data from the File:

```python
# Download the file

!wget https://s3.amazonaws.com/weclouddata/datasets/genai/python_fundamentals/data.txt
```

```python
# Read the contents of the file
with open('data.txt', 'r') as file:
    data = file.read()

print(data)
```

### Parsing Data into Lists:

```python
# Split the lines into a list
lines = data.split('\n')

# Remove the header and empty last line
header = lines[0].split(', ')
lines = [line.split(', ') for line in lines[1:-1]]

print("Header:", header)
print("Data:", lines)
```

### Filtering Data Using if/else:

```python
# Filter people with age greater than 25
filtered_data = [line for line in lines if int(line[1]) > 25]

print("People with Age > 25:", filtered_data)
```

### Updating Data:

```python
# Update Charlie's age to 23
for line in lines:
    if line[0] == 'Charlie':
        line[1] = '23'

print("Updated Data:", lines)
```

### Creating New Data:

```python
# Add a new person to the data
new_person = ['Frank', '27', 'B']
lines.append(new_person)

print("Updated Data with New Person:", lines)
```
