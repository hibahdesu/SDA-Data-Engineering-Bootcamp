```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("telecom.csv")
data.head()
```

### 1. Create a pivot table for the follwing:
>    - The average `MonthlyCharges` per `InternetService` for males and females.
>    - The maximum `MonthlyCharges` per `Contract` for Seniors / Non-seniors.
>    - The sum of the `MonthlyCharges` per `PaymentMethod` for Churn / Non-churn customers
>    - Also, try using the tradtional group by aggregations combined with the unstack() method to create the above tables.

```python
## Average MonthlyCharges per InternetService for males and females
```

```python
# METHOD 1: Traditional way of pivot
data.groupby(['InternetService', 'gender'])[['MonthlyCharges']].aggregate('mean').unstack()
```

```python
# METHOD 2: Pandas pivot_table
data.pivot_table('MonthlyCharges', index='InternetService', columns='gender', aggfunc='mean')
```

```python
## Maximum MonthlyCharges per Contract for Seniors/Non-seniors
```

```python
# METHOD 1: Traditional way of pivot
data.groupby(['Contract', 'SeniorCitizen'])[['MonthlyCharges']].aggregate('max').unstack()
```

```python
# METHOD 2: Pandas pivot_table
data.pivot_table('MonthlyCharges', index='Contract', columns='SeniorCitizen', aggfunc='max')
```

```python
## Sum of the MonthlyCharges per PaymentMethod for Churn/Non-churn customers
```

```python
# METHOD 1: Traditional way of pivot
data.groupby(['PaymentMethod', 'Churn'])[['MonthlyCharges']].aggregate('sum').unstack()
```

```python
# METHOD 2: Pandas pivot_table
data.pivot_table('MonthlyCharges', index='PaymentMethod', columns='Churn', aggfunc='sum')
```

### 2. Create a column named `InternetService Average`:
>    - For each customer, this value will be the overall average monthly charge for their respective internet service. (Overall average DSL monthly charge, Overall average Fibre Optic monthly charge, ...)
>    - Create a new column containing the ratio between the customers `MonthlyCharges` and the `InternetService Average`

```python
data.head()
```

```python
col_of_interest = ['customerID', 'InternetService', 'MonthlyCharges', 'InternetServiceAverage']
```

```python
# Calculating the average monthly charges for each internet service
# Similar to using a window function in SQL
    # i.e. AVG(MonthlyService) OVER (PARTITION BY InternetService)
data['InternetServiceAverage'] = data.groupby('InternetService')['MonthlyCharges'].transform(np.mean)
data[col_of_interest]
```

```python
data['ratio'] = data['MonthlyCharges'] / data['InternetServiceAverage']
data[col_of_interest + ['ratio']] # Just selecting the columns we're interested in
```

### 6. Create a new column `normalized_salary`:
>    - This column will return the following for each employee: `employees salary - average salary / standard deviation`

```python
# recall axis=0 means column-wise
salaries['avg_salary'] = round(salaries['salary'].mean(axis=0))
salaries
```

```python
# Standard score, z = (X - µ) / σ
    # z is negative when the raw score is below the mean,
    # positive when above
salaries['normalized_salary'] = (salaries['salary'] - salaries['avg_salary']) / salaries['salary'].std()
salaries
```

### 11. Return the job titles where the average salary of females were higher than males.

```python
# Here, we'll consider each record in the salaries table to be an individual data point
# E.g. A male Staff employee may have multiple records in the salaries table due to
# getting raises while working in the same position (each of these records would be
# counted individually)
```

```python
# 1. Merge employees and salaries together so we can get the gender & salary info
q11 = pd.merge(employees, salaries, on='emp_no')[['emp_no', 'gender', 'salary', 'from_date', 'to_date']]
q11
```

```python
# Note: titles table shows the title of an employee over the entire duration they held
# that title (unlike the salaries table, which has a new row every time their salary changes)

# Example (in cells below):
```

```python
# titles[titles['emp_no'] == 499998]
```

```python
# q11[q11['emp_no'] == 499998]
```

```python
# 2. We're going to merge the table we created above, q11, with titles to get the title info
# For speed/simplicity I'm creating a copy with only the columns we need
titles_trunc = titles[['emp_no', 'title', 'from_date', 'to_date']].copy()
titles_trunc
```

```python
# If q11's from_date and to_date are within titles' from_date and to_date, then assign the title
# E.g. emp_no 499998's from_date: 1993-12-27,
#                      to_date: 1994-12-27,
#                      salary: 40000
# fall within the time when they were working as a Staff (1993-12-27 -> 1998-12-27)
```

```python
# 3. Join the two tables so we can later check if the from_date_x & to_date_x from q11
# is within the from_date_y & to_date_y from titles_trunc
q11_with_titles = pd.merge(q11, titles_trunc, on='emp_no')
q11_with_titles
```

```python
# 4. Filter the rows from the above table to get an employee's correct title during the time given
# Note: the .query expression used below is equivalent to..
    # q11_with_titles[(q11_with_titles['from_date_x'] >= q11_with_titles['from_date_y']) &
    # (q11_with_titles['to_date_x'] <= q11_with_titles['to_date_y'])]

desired_rows = q11_with_titles.query('from_date_x >= from_date_y and to_date_x <= to_date_y')
desired_col = ['emp_no', 'gender', 'salary', 'title', 'from_date_x', 'to_date_x']

q11_final = desired_rows[desired_col]
q11_final
```

```python
# 5. Get job titles where the average salary of females were higher than males
    # Use pivot table to find what these job titles are
q11_pivot = round(q11_final.pivot_table('salary', index='title', columns='gender', aggfunc='mean'))
q11_pivot
```

```python
q11_pivot.query('F > M') # same as: q11_pivot[q11_pivot['F'] > q11_pivot['M']]

## ANS: 'Staff' is the only job title where avg salary of females are higher than males
```
